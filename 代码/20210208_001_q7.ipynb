```python
import pandas as pd
data=pd.read_csv("TempData/q3_cleaned_withtype_data.csv",encoding="utf-8")
```

```python
from nltk.tokenize import word_tokenize
import nltk as nk
def det(result):
    '''输入一段文本，返回单词列表'''
    result=word_tokenize(result)
    value=['*',',','!','.',"'",'+','-','/','\\',';',':','?','(',')','<','>']
    j = 0
    for i in range(len(result)):
        if result[j] in value:
            result.pop(j)
        else:
            j += 1
    return result
```

```python
label1='1921-1930'
group1=data[(data['year']>=1921)&(data['year']<1930)]['song_title (censored)']
text=group1.tolist()
text1="".join(text)
text1=det(text1)
freq = nk.FreqDist(text1)
standard_freq=freq.most_common(5)
print(standard_freq)
#绘图函数为这些词频绘制一个图形
freq.plot(20, cumulative=False)
```
