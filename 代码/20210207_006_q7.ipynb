```python
import pandas as pd
data=pd.read_csv("TempData/q3_cleaned_withtype_data.csv",encoding='utf-8')
```

```python
import matplotlib.pyplot as plt
group=data.groupby(['type'])
i=1
plt.figure(figsize=(10,10))
for a,b in group:
    ax=plt.subplot(4,5,i)
    print(a)
    c=b.groupby(['year'])['id_count'].count()
    c=pd.DataFrame(c)
    year=list(c.index)
    count=c['id_count'].tolist()
    i=i+1
    ax.plot(year,count,'r',label=a)
    plt.legend()
plt.show()
```
